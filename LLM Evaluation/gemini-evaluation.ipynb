```python
from rouge import Rouge

# Assuming you have two lists of summaries: one for reference summaries and one for generated summaries
reference_summaries = reference  # Add your 50 reference summaries here
generated_summaries = answers  # Add your 50 generated summaries here

# Initialize the ROUGE scorer
rouge = Rouge()

# Calculate ROUGE scores for each pair
all_scores = []

for reference, generated in zip(reference_summaries, generated_summaries):
    scores = rouge.get_scores(generated, reference)
    all_scores.append(scores[0])

# If you want to see the scores for each pair
for i, scores in enumerate(all_scores, start=1):
    print(f"Pair {i}: ROUGE-1: {scores['rouge-1']['f']}, ROUGE-2: {scores['rouge-2']['f']}, ROUGE-L: {scores['rouge-l']['f']}")

# If you want to calculate average scores
avg_scores = {
    'rouge-1': {'f': 0, 'p': 0, 'r': 0},
    'rouge-2': {'f': 0, 'p': 0, 'r': 0},
    'rouge-l': {'f': 0, 'p': 0, 'r': 0},
}

for scores in all_scores:
    for key in avg_scores.keys():
        avg_scores[key]['f'] += scores[key]['f']
        avg_scores[key]['p'] += scores[key]['p']
        avg_scores[key]['r'] += scores[key]['r']

for key in avg_scores.keys():
    avg_scores[key]['f'] /= len(all_scores)
    avg_scores[key]['p'] /= len(all_scores)
    avg_scores[key]['r'] /= len(all_scores)

print("\nAverage scores:")
for key, value in avg_scores.items():
    print(f"{key}: F1: {value['f']}, Precision: {value['p']}, Recall: {value['r']}")
```

```python
safety_settings=[
    {
        "category": "HARM_CATEGORY_HARASSMENT",
        "threshold": "BLOCK_NONE",
    },
    {
        "category": "HARM_CATEGORY_HATE_SPEECH",
        "threshold": "BLOCK_NONE",
    },
    {
        "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
        "threshold": "BLOCK_NONE",
    },
    {
        "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
        "threshold": "BLOCK_NONE",
    },
    ]
```

```python
import time
import csv

answers = []
for i in range(0, 10):
    model = genai.GenerativeModel("gemini-1.5-pro-latest", safety_settings=safety_settings)
    chat = model.start_chat(history=[])

    try:
        response = chat.send_message(promt.format(text=df['text_en'].iloc[i]))

        if response:
            output = response.text
    # Handle the blocked prompt exception
    except genai.types.generation_types.BlockedPromptException as e:
        print(f"{i} Prompt blocked due to: {e}")
        output = "Empty"

    time.sleep(0.25) # you can skip this line
    answers.append(output)
    # Open the CSV file in append mode to add each response at the end of the file
    with open('output_responses.csv', 'a', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow([answers[i]])  # Writing each response in a new row
    print(f"Response {i} has been saved to 'output_responses.csv'.")
    print(answers[i])

    print("All responses have been saved.")
```
